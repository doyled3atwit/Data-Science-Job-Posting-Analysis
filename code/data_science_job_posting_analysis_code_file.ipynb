```python
# imports 

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# import data 
# B for benefits 
B_NA = pd.read_csv('benefits.csv')

# drop empty rows 
B_unnamed = B_NA.dropna()

B = B_unnamed.rename(columns={'Unnamed: 0': 'ID'})

# 558 rows after clean 
# this data only has entries of 0 = no, 1 = yes 

# display first 5 rows of data 
B.head()

#display(B)
```

```python
# counting benefits per col 
# out of 558 IDs 
count_yes_per_col = B.eq(1).sum().sort_values(ascending=False)
count_no_per_col = B.eq(0).sum().sort_values(ascending=False)

print(count_yes_per_col)
print()
#print(count_no_per_col)

top_5_benefits = count_yes_per_col.head(5)


mylabels = ['Health Insurance : 508','Dental Insurance : 354', '401(k) : 248', 'Paid Time Off : 155', 'Vision Insurance : 110'] 

plt.figure(figsize=(15,8))
#plt.hist(top_5_benefits, labels = mylabels , textprops = {'fontsize' : 13})
#plt.title("The 5 Most Common Benefits Amoung Data Science Jobs", fontsize = 16)

plt.bar(mylabels, top_5_benefits, color='violet')

# Adding labels and title
plt.xlabel('\nBenefits', fontsize = 14)
plt.ylabel('Number of Job Posting with Benefit\n', fontsize= 14)
plt.title("The 5 Most Common Benefits From 558 Data Science Jobs Postings", fontsize = 16)

plt.tick_params(axis='y', which='major', labelsize=12)  # Adjust the label size
plt.tick_params(axis='y', which='major', length=8, width=2)  # Adjust the tick size
plt.tick_params(axis='x', which='major', labelsize=12)  # Adjust the label size

plt.show()

```

```python
# import data 
# Q for qualifications 
# 1287 rows X 363 col 
Q_unnamed = pd.read_csv('qualifications.csv')

Q = Q_unnamed.rename(columns={'Unnamed: 0': 'ID'})


# display first 5 rows of data 
Q.head()

# display(Q)
```

```python
# counting the number of jobs reqiring a masters degree
# 581 / 1279 do not require masters 
# 698 / 1279 do reqire masters 

Master = Q.groupby("Master's degree").agg({'ID': 'count'})
print(Master)

# pie chart 

plt.figure(figsize=(8, 8)) # plot size 
#sns.set_palette('pastel') # choose color 
plt.pie(Master['ID'], # what to plot 
        labels = ['Not Required', 'Required'], 
        colors = ['pink', 'lightblue'],
        autopct=lambda p: '{:.0f} ({:.1f}%)'.format(p * sum(Master['ID']) / 100, p), # displays the count and % 
        startangle = 130 , # changes how the pieces of the pie are positioned / start ( think 360 degreee circle)
        textprops={'fontsize': 15}) # font size of labels 
plt.title("Is a Master's Degree a Vital Qualification for Data Science Positions ?", fontsize = 15 ) # title and its font size 

```

```python
# which SELECTED skills are most important 

# NOTE : THIS IS ONLY COMPARING SELECTED COLUMNS NOT ALL DATA  

Skills = Q[['R', 'Research', 'Machine learning', 'Analysis skills', 'Communication skills', 'Python']]
count_ones_skills = Skills.eq(1.0).sum()
print(count_ones_skills)

plt.figure(figsize=(8,8))

plt.pie(count_ones_skills , 
       labels = ['R', 'Research', 'Machine Learning', 'Analysis Skills', 'Communication Skills', 'Python'],
       textprops = {'fontsize' : 15})

plt.title( 'What skills are most important amoung these selected skills ?', fontsize = 20)
           
plt.show()
```

```python
allSkills = Q[Q.columns]
count_ones_allSkills = allSkills.eq(1.0).sum().sort_values(ascending = False)

# print(count_ones_allSkills)

count_ones_allSkills_10 = count_ones_allSkills.head(10)

print(count_ones_allSkills_10)

plt.figure(figsize=(8,8))

plt.pie(count_ones_allSkills_10 , 
       labels = count_ones_allSkills_10.index ,
       textprops = {'fontsize' : 15})

plt.title( 'What skills are most important ?', fontsize = 20)
           
plt.show()
```

```python
# common coding landuages 
codeLang = Q[['R', 'Python', 'C' ,'C++', 'SQL', 'Java' ]]
count_ones_codes = codeLang.eq(1.0).sum()
print(count_ones_codes)

plt.figure(figsize=(8,8))

plt.pie(count_ones_codes , 
       labels = ['R', 'Python','C', 'C++', 'SQL', 'Java' ],
       textprops = {'fontsize' : 15})

plt.title( 'What coding language is most desired ?', fontsize = 20)
           
plt.show()
```

```python
# import data csv file 
# T for title location company salary 
T_unnamed = pd.read_csv('title_location_company_salary.csv')

# this data has mostly strings and float salary 

T_NA = T_unnamed.rename(columns={'Unnamed: 0': 'ID'})

TD = T_NA.dropna()

T = TD.drop_duplicates()

# display first 5 rows of data 
T.head()

#display(T)
# useful for salary 
# T.describe()
```

```python
#finding highest salary
highest_salaries = T.sort_values(by='Salary', ascending=False)

top_10_salaries = highest_salaries.head(10)

print(top_10_salaries)
print()

#make pie chart of location and companys 

company_counts = highest_salaries['Company'].head(20).value_counts()

plt.figure(figsize=(12,8))
plt.pie(company_counts, labels = company_counts.index, textprops = {'fontsize' : 14})
plt.title("Companies Offering the Top 20 Highest Salaries in California \n", fontsize = 16)

plt.show()
print()

locations_counts = highest_salaries['Location'].head(20).value_counts()

plt.figure(figsize=(12,8))
plt.pie(locations_counts, labels = locations_counts.index, textprops = {'fontsize' : 14}) 
plt.title("Locations with the Top 20 Highest Salaries in California \n", fontsize = 16)

plt.show 
```

```python
#counting occurrences of each value in the Salary column  
value_counts_salary = T['Salary'].value_counts().sort_values(ascending=False) 
top_value_counts = value_counts_salary.head(10)
print(top_value_counts, '\n')

print(T['Salary'].describe(), '\n')
print(T['Location'].describe(), '\n')
```
